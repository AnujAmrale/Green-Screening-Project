```python
!pip install pandas
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
df = pd.read_csv("/content/drive/My Drive/Colab Notebooks/global_warming_dataset.csv")
print(df.head())

import seaborn as sns
import matplotlib.pyplot as plt

print("Dataset Info:\n", df.info())
print("\nMissing Values:\n", df.isnull().sum())
print("\nDuplicate Rows:", df.duplicated().sum())

df.dropna(inplace=True)
print("\nAfter Cleaning:\nMissing Values:\n", df.isnull().sum())
print("\nDataset Summary:\n", df.describe())

# Visualization 1: Temperature Anomalies Over Time
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Year", y="Temperature_Anomaly", estimator="mean", ci=None)
plt.xlabel("Year")
plt.ylabel("Average Temperature Anomaly (Â°C)")
plt.title("Global Temperature Anomalies Over Time")
plt.grid(True)
plt.show()

# Visualization 2: CO2 Emissions vs GDP
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="GDP", y="CO2_Emissions", alpha=0.5)
plt.xscale("log")
plt.xlabel("GDP (Log Scale)")
plt.ylabel("CO2 Emissions")
plt.title("GDP vs CO2 Emissions")
plt.grid(True)
plt.show()
```
